```python
# Преобразователь таблиц из https://rise-of-cultures.fandom.com/wiki/World_Wonders/Contribution_Rewards

table_text = '''
1	Research.png 5
Blueprint-Temple of Artemis.png 1	-	-	-
2	Research.png 10
Blueprint-Temple of Artemis.png 2	Research.png 5
Blueprint-Temple of Artemis.png 1	-	-
3	Research.png 10
Blueprint-Temple of Artemis.png 3	Research.png 5
Blueprint-Temple of Artemis.png 2	Research.png 3
Blueprint-Temple of Artemis.png 1	-
4	Research.png 15
Blueprint-Temple of Artemis.png 3	Research.png 10
Blueprint-Temple of Artemis.png 2	Research.png 5
Blueprint-Temple of Artemis.png 1	-
5	Research.png 20
Blueprint-Temple of Artemis.png 3	15
2

5
1

6	Research.png 30
Blueprint-Temple of Artemis.png 3			
7	Research.png 35
Blueprint-Temple of Artemis.png 3			
8	Research.png 45
Blueprint-Temple of Artemis.png 3	Research.png 25
Blueprint-Temple of Artemis.png 2	Research.png 15
Blueprint-Temple of Artemis.png 1	Research.png 5
'''

tex_table = '''
\\begin{longtable}[c]{|c|c|c|c|}
    \\hline
    Уровень & Место & Колб & Чертежей \\\\\\hline\\endhead
'''    

lines = []
for line in table_text.split('\n'):
    line = line.replace('\t', ' ')
    line = line.split(' ')
    if len(line) > 1:
        if line[0].isnumeric():
            lines.append(line)
        else:
            lines[-1].extend(line)
        
for line in lines:
    level = int(line[0])
    prizes = []
    prize = []
    for element in line[2:]:
        if element.isnumeric():
            if len(prize) == 0:
                prize.append(int(element))
            else:
                prize.append(int(element))
                prizes.append(prize)
                prize = []
    if len(prize) == 1:
        prize.append(0)
        prizes.append(prize)
    tex_table += f'    \\multirow{{{len(prizes)}}}{{*}}{{{level}}} & 1 & {prizes[0][0]} & {prizes[0][1]}'
    for prize, index in zip(prizes[1:], range(2, len(prizes) + 1)):
        tex_table += f' \\\\\\cline{{2-4}}\n    & {index} & {prize[0]} & {prize[1]}'
    tex_table += ' \\\\\\hline\n'
    
        
tex_table += '\\end{longtable}'
print(tex_table)
```
